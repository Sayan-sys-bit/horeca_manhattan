```python
# Load the dataset
df = pd.read_csv('Zomato Dataset.csv')
sns.set_style("whitegrid")
```

```python
top_cuisines = df['Cuisine'].value_counts().nlargest(10)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_cuisines.index, y=top_cuisines.values, palette="viridis")
plt.title('Top 10 Most Popular Cuisines (Item Count)', fontsize=16)
plt.xlabel('Cuisine', fontsize=12)
plt.ylabel('Number of Items Listed', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_cuisines_popularity.png')
```

```python
# 2.2. Price Distribution and High-Value Segment
plt.figure(figsize=(10, 6))
sns.histplot(df['Log_Price'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Log-Transformed Item Prices', fontsize=16)
plt.xlabel('Log(Price)', fontsize=12)
plt.ylabel('Density/Count', fontsize=12)
plt.axvline(df[df['Is_Expensive'] == 1]['Log_Price'].min(), color='red', linestyle='--', label='Boundary of Expensive')
plt.legend()
plt.tight_layout()
plt.savefig('log_price_distribution.png')
```

```python
# 2.3. Key Insight Derivation (Cuisine Rating vs. Price)
cuisine_summary_A1 = df.groupby('Cuisine').agg(
    Avg_Price=('Prices', 'mean'),
    Avg_Rating=('Average_Rating', 'mean'),
    Item_Count=('Item_Name', 'count')
).reset_index()
```

```python
# Filter for the top 15 most frequent cuisines for clarity
top_cuisine_names = top_cuisines.index.tolist()
plot_data_A1 = cuisine_summary_A1[cuisine_summary_A1['Cuisine'].isin(top_cuisine_names)]
```

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Avg_Price', y='Avg_Rating', data=plot_data_A1,
                size='Item_Count', sizes=(50, 500),
                hue='Cuisine', legend=False, alpha=0.7)
for line in range(0, plot_data_A1.shape[0]):
    plt.text(plot_data_A1.Avg_Price.iloc[line]*1.01, plot_data_A1.Avg_Rating.iloc[line],
             plot_data_A1.Cuisine.iloc[line], horizontalalignment='left', size='large', color='black', weight='semibold')
    plt.title('Cuisine Value Map: Average Price vs. Average Rating (Top 15 Cuisines)', fontsize=16)
plt.xlabel('Average Price (INR)', fontsize=12)
plt.ylabel('Average Rating (0-5)', fontsize=12)
plt.tight_layout()
plt.savefig('cuisine_value_map.png')
```

```python
# --- 3. STRATEGIC AREA 2: LOCATION AND POPULARITY DRIVERS ---

# 3.1. City Performance
city_summary = df.groupby('City').agg(
    Avg_Rating=('Average_Rating', 'mean'),
    Avg_Price=('Prices', 'mean'),
    Total_Items=('Item_Name', 'count')
).reset_index()
```

```python
top_cities = city_summary.sort_values(by='Total_Items', ascending=False).head(5)
```

```python
plt.figure(figsize=(10, 6))
# Compare Average Rating and Average Price across top cities
top_cities_melted = top_cities.melt(id_vars='City', value_vars=['Avg_Rating', 'Avg_Price'],
                                    var_name='Metric', value_name='Value')

sns.barplot(x='City', y='Value', hue='Metric', data=top_cities_melted, palette='coolwarm')
plt.title('Average Rating and Price Comparison for Top 5 Cities', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('city_performance_comparison.png')
#
```

```python
# 3.2. Bestseller Impact Analysis
bestseller_impact = df.groupby('Is_Bestseller')[['Average_Rating', 'Prices', 'Total_Votes']].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x='Is_Bestseller', y='Average_Rating', data=bestseller_impact, palette='Paired')
plt.title('Average Rating: Bestseller vs. Non-Bestseller', fontsize=16)
plt.xlabel('Is Bestseller (0=No, 1=Yes)', fontsize=12)
plt.ylabel('Average Item Rating', fontsize=12)
plt.xticks([0, 1], ['Non-Bestseller', 'Bestseller'])
plt.tight_layout()
plt.savefig('bestseller_rating_impact.png')
#
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Log_Price_per_Vote'], kde=True, bins=30, color='coral')
median_lpv = df['Log_Price_per_Vote'].median()
plt.axvline(median_lpv, color='red', linestyle='--', label=f'Median Value (Log P/V): {median_lpv:.2f}')
plt.title('Distribution of Log(Price per Vote) - The Value Metric', fontsize=16)
plt.xlabel('Log(Price / Total Votes)', fontsize=12)
plt.ylabel('Density/Count', fontsize=12)
plt.legend()
plt.tight_layout()
plt.savefig('value_metric_distribution.png')
```

```python
# Calculate 'Price_per_Vote' and 'Log_Price_per_Vote'
df['Price_per_Vote'] = df['Prices'] / df['Total_Votes']
# Add a small constant to handle cases where Total_Votes is 0 to avoid division by zero issues and log(0)
df['Log_Price_per_Vote'] = np.log(df['Price_per_Vote'] + 1e-9)
```

```python
# 2. New Plot 2: Deep Dive on Value (Cuisine vs. Price-per-Vote)
top_cuisines = df['Cuisine'].value_counts().nlargest(8).index.tolist()
df_top_cuisine = df[df['Cuisine'].isin(top_cuisines)]

plt.figure(figsize=(12, 7))
sns.boxplot(x='Cuisine', y='Log_Price_per_Vote', data=df_top_cuisine, palette='tab10')
plt.title('Value for Money (Log P/V) by Top Cuisine', fontsize=16)
plt.xlabel('Cuisine', fontsize=12)
plt.ylabel('Log(Price / Total Votes) [Lower is Better Value]', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('cuisine_value_comparison.png')
```

```python
# 3. New Plot 3: Key Drivers Analysis (Correlation Heatmap)
# Select key numerical columns for correlation analysis
corr_cols = ['Average_Rating', 'Total_Votes', 'Prices', 'Dining_Rating', 'Delivery_Rating',
             'Dining_Votes', 'Delivery_Votes', 'Price_per_Vote', 'Log_Price']

# Calculate the correlation matrix
corr_matrix = df[corr_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm',
            cbar_kws={'label': 'Pearson Correlation Coefficient'})
plt.title('Correlation Heatmap of Key Performance Drivers', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('key_drivers_correlation_heatmap.png')
```

```python
# 4. New Plot 4: Restaurant Scale vs. Quality (Popularity vs. Rating)
# Aggregate data at the Restaurant level (since 'Avg_Rating_Restaurant' and 'Restaurant_Popularity' are already pre-calculated, we can use the original df for the scatter)
# Filter to unique restaurant rows to avoid overplotting and misinterpretation
restaurant_agg = df[['Restaurant_Name', 'Avg_Rating_Restaurant', 'Restaurant_Popularity']].drop_duplicates()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Restaurant_Popularity', y='Avg_Rating_Restaurant', data=restaurant_agg,
                alpha=0.6, color='darkgreen')
plt.title('Restaurant Quality vs. Scale (Item Count)', fontsize=16)
plt.xlabel('Restaurant Scale (Number of Items Listed)', fontsize=12)
plt.ylabel('Average Restaurant Rating', fontsize=12)
plt.tight_layout()
plt.savefig('restaurant_scale_vs_quality.png')

print("All four detailed plots have been generated and saved.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
try:
    df = pd.read_csv('Zomato Dataset.csv')
    sns.set_style("whitegrid")
except Exception as e:
    print(f"Error loading data: {e}")
    df = None

if df is not None:
    # Prepare aggregated city data for Plot 5 and 9
    city_data = df[['City', 'Avg_Rating_City', 'Avg_Price_City']].drop_duplicates().sort_values(by='Avg_Rating_City', ascending=False)

    # ---------------------------------------------------------------------
    # 1. Plot 5: Relationship between Avg_Price_City and Avg_Rating_City
    # ---------------------------------------------------------------------
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Avg_Price_City', y='Avg_Rating_City', data=city_data,
                    size='Avg_Rating_City', sizes=(100, 500),
                    hue='City', legend=False, alpha=0.7)

    # Annotate top/bottom performing cities for context
    top_city = city_data.head(1).iloc[0]
    bottom_city = city_data.tail(1).iloc[0]

    plt.text(top_city['Avg_Price_City'], top_city['Avg_Rating_City']-0.01, top_city['City'],
             weight='bold', color='darkgreen')
    plt.text(bottom_city['Avg_Price_City'], bottom_city['Avg_Rating_City']+0.01, bottom_city['City'],
             weight='bold', color='red')

    plt.title('City Market Positioning: Average Price vs. Average Rating', fontsize=16)
    plt.xlabel('Average Item Price in City (INR)', fontsize=12)
    plt.ylabel('Average Item Rating in City', fontsize=12)
    plt.tight_layout()
    plt.savefig('city_market_positioning.png')

    # ---------------------------------------------------------------------
    # 2. Plot 6: Item Price Distribution by Best_Seller Status
    # ---------------------------------------------------------------------

    # Clean/map Best_Seller to categories
    df['Best_Seller_Group'] = df['Best_Seller'].replace({'BESTSELLER': 'BESTSELLER', 'MUST TRY': 'MUST_TRY', 'NONE': 'NONE'})

    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Best_Seller_Group', y='Prices', data=df,
                order=['BESTSELLER', 'MUST_TRY', 'NONE'], palette='Set2', showfliers=False) # showfliers=False to ignore extreme outliers

    median_bestseller_price = df[df['Best_Seller_Group'] == 'BESTSELLER']['Prices'].median()

    plt.title('Item Price Distribution by Popularity Status', fontsize=16)
    plt.xlabel('Popularity Tag', fontsize=12)
    plt.ylabel('Item Price (INR)', fontsize=12)
    plt.text(0.5, median_bestseller_price, f'Bestseller Median: {median_bestseller_price:.2f}',
             ha='center', va='bottom', color='red', weight='bold')
    plt.tight_layout()
    plt.savefig('price_by_bestseller_status.png')

    # ---------------------------------------------------------------------
    # 3. Plot 7: Dining vs. Delivery Ratings by Cuisine
    # ---------------------------------------------------------------------
    top_cuisines = df['Cuisine'].value_counts().nlargest(8).index.tolist()
    cuisine_channel_rating = df[df['Cuisine'].isin(top_cuisines)].groupby('Cuisine')[['Dining_Rating', 'Delivery_Rating']].mean().reset_index()

    cuisine_channel_rating_melted = cuisine_channel_rating.melt(id_vars='Cuisine',
                                                                value_vars=['Dining_Rating', 'Delivery_Rating'],
                                                                var_name='Channel',
                                                                value_name='Avg_Rating')

    plt.figure(figsize=(12, 6))
    sns.barplot(x='Cuisine', y='Avg_Rating', hue='Channel', data=cuisine_channel_rating_melted,
                palette={'Dining_Rating': 'skyblue', 'Delivery_Rating': 'salmon'})
    plt.title('Channel Performance Gap: Dining vs. Delivery Ratings by Top Cuisine', fontsize=16)
    plt.xlabel('Cuisine', fontsize=12)
    plt.ylabel('Average Channel Rating', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Channel')
    plt.tight_layout()
    plt.savefig('channel_gap_by_cuisine.png')

    # ---------------------------------------------------------------------
    # 4. Plot 8: Price vs. Total Votes with Is_Highly_Rated Hue
    # ---------------------------------------------------------------------

    # Downsample for cleaner scatter plot due to large dataset size
    df_sample = df.sample(n=10000, random_state=42)

    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='Prices', y='Total_Votes', data=df_sample,
                    hue='Is_Highly_Rated', palette={0: 'gray', 1: 'gold'}, alpha=0.6)

    plt.title('Revenue Potential: Price vs. Votes, Colored by High Rating ($\geq 4.0$)', fontsize=16)
    plt.xlabel('Item Price (INR)', fontsize=12)
    plt.ylabel('Total Votes (Popularity Proxy)', fontsize=12)
    plt.xlim(0, df['Prices'].quantile(0.95)) # Focus on 95% of data to manage outliers
    plt.ylim(0, df['Total_Votes'].quantile(0.95))
    plt.legend(title='Highly Rated')
    plt.tight_layout()
    plt.savefig('price_votes_highly_rated.png')

    # ---------------------------------------------------------------------
    # 5. Plot 9: Price Disparity Gap (Restaurant Avg Price vs. City Avg Price)
    # ---------------------------------------------------------------------

    # Aggregate data at the Restaurant level, calculating the price difference
    restaurant_gap_df = df[['Restaurant_Name', 'Avg_Rating_Restaurant', 'Avg_Price_Restaurant', 'Avg_Price_City']].drop_duplicates()
    restaurant_gap_df['Price_Premium_Gap'] = restaurant_gap_df['Avg_Price_Restaurant'] - restaurant_gap_df['Avg_Price_City']

    # Create rating groups for hue
    restaurant_gap_df['Rating_Group'] = pd.cut(restaurant_gap_df['Avg_Rating_Restaurant'],
                                               bins=[0, 3.5, 4.0, 5.0],
                                               labels=['Low (<3.5)', 'Medium (3.5-4.0)', 'High (>4.0)'])

    plt.figure(figsize=(10, 6))
    sns.histplot(x='Price_Premium_Gap', data=restaurant_gap_df, hue='Rating_Group',
                 multiple='stack', bins=30, palette='viridis', edgecolor='black')

    plt.axvline(0, color='red', linestyle='--', linewidth=2, label='City Price Parity')
    plt.title('Competitive Price Positioning vs. Market Average', fontsize=16)
    plt.xlabel('Price Premium Gap (Restaurant Avg Price - City Avg Price)', fontsize=12)
    plt.ylabel('Number of Restaurants', fontsize=12)
    plt.legend(title='Avg Restaurant Rating')
    plt.tight_layout()
    plt.savefig('price_premium_gap_distribution.png')

    print("Five new detailed plots have been generated and saved.")
else:
    print("Cannot proceed with plotting as the DataFrame is not loaded.")
```
